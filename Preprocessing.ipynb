### convert age column from days to year

df['age']=df['age']//365
df

### Remove the rows containing unrealistic systolic (ap_hi) and diastolic (ap_lo) blood pressure values from the DataFrame by keeping:

* ap_hi between 70 and 250
* ap_lo between 40 and 150

```python
df = df[(df['ap_hi'] >= 70) & (df['ap_hi'] <= 250)]
df = df[(df['ap_lo'] >= 40) & (df['ap_lo'] <= 150)]
 # Also ensure ap_hi is always greater than ap_lo
df = df[df['ap_hi'] >= df['ap_lo']]
```

### How many patients are smokers?

```python
df['cardio'].value_counts()
```

### Peoples having high cholesterol

```python
# cholesterol-> 2 = High
len(df[df['cholesterol'] == 2]) 
```

### If the DataFrame contains weight and height columns, calculate the BMI

```python
if 'weight' in df.columns and 'height' in df.columns:
    df['bmi'] = df['weight'] / ((df['height'] / 100)**2)
    df['bmi'] = df['bmi'].round(1)

# Drop after BMI is made
df = df.drop(columns=['height', 'weight'], errors='ignore')

df
```

### Filter the DataFrame to keep only those records where the BMI value is between 20 and 40.

```python
df=df[(df['bmi']>20)&(df['bmi']<40)]
```

### Scatter Plot Analysis
Create a Scatter Plot to analyze the relationship between `Height` and `weight`.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='ap_hi',y='ap_lo',data=df ,alpha=0.5)
plt.title('Systolic blood pressure vs Diastolic blood pressure')
plt.xlabel('Systolic blood pressure')
plt.ylabel('Diastolic blood pressure')
plt.show()
```

### Plot histogram of age values using matplotlib.

```python
import matplotlib.pyplot as plt

plt.hist(df["age"], bins=30)
plt.xlabel("Age (days)")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

### Plot count of cholesterol levels.

```python
df["cholesterol"].value_counts().plot(kind="bar")
plt.xlabel("Cholesterol Level")
plt.ylabel("Count")
plt.title("Cholesterol Levels Distribution")
plt.show()
```
